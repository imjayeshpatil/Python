```python
#one plan multiple output genrate
#plot merging in matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2*np.pi,0.02)#sample pass though arrange

y1 = np.sin(x)

y2 = np.cos(x)

plt.plot(x,y1,label='Sin')

plt.plot(x,y2,label='Cos')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend() #show label
plt.title('Sin & Cosine Function')

plt.show()
```
