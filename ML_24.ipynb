```python
#pie char is used for distribute data in Percentage
import matplotlib.pyplot as plt

activities = ['eat','sleep','work','play']

slices = [3,7,8,6]

color = ['r','g','m','b']

plt.pie(slices, labels=activities, colors=color,startangle=90,shadow=True,explode=(0.2,0,0,0),autopct='%1.2f%%') 
#add start point easy to understand use startangle=90,3D view add Shadow use shadow=True,use explode for cut slice
#get value in percentage use autopct='%1.2f%%'

plt.legend() 
plt.show()
```
