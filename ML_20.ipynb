```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.linspace(0.1,5.0) #linspace is an in-built function in Python's NumPy library. It is used to create an evenly spaced sequence in a specified interval.
x2 = np.linspace(0.0,2.0) # linspace is a tool in Python for creating numeric sequences

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)

y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('Subplot-1')
plt.xlabel('x1')
plt.ylabel('Amp(y1)')

plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('subplot-2')
plt.xlabel('x2')
plt.ylabel('Amp(y2)')
plt.show()
```
